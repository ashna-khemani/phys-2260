# Plotting polar equations

```python
from numpy import cos, sin, exp, linspace, pi
from pylab import plot, show, xlabel, ylabel, subplot, legend

# Convert polar coordinates to Cartesian coordinates
def cartesian(r, theta):
    x = r*cos(theta)
    y = r*sin(theta)
    return [x,y]

# First function
theta = linspace(0,10*pi, 1000)
r = theta**2
x1, y1 = cartesian(r,theta)

# Second function
theta = linspace(0, 24*pi, 1000)
r = exp(cos(theta)) - 2*cos(4*theta) + (sin(theta/12))**5
x2, y2 = cartesian(r, theta)

# Third function
theta = linspace(0,2*pi, 10000)
r = 2*(1+cos(theta))
x3, y3 = cartesian(r, theta)

# Plot results
plot(x1, y1, "r-", label=r"$r = \theta^2$")
legend()
xlabel("x")
ylabel("y")
show()

plot(x2, y2, "b-", label=r"$r = e^{\cos\ \theta} - 2\cos\ 4\theta + \sin^5\ \theta/12$")
legend()
xlabel("x")
ylabel("y")
show()

plot(x3, y3, "g--", label=r"$2(1 + \cos\ \theta)$")
legend()
xlabel("x")
ylabel("y")
show()

```

## Density plot of $z(x,y) = \cos x \sin y$

```python
from numpy import zeros, sin, cos, pi
from pylab import imshow, show, colorbar, xlabel, ylabel

points = 500        # Number of grid points along each side
xmax = 10
xmin = -xmax
ymax = 10
ymin = -ymax

dx = (xmax - xmin)/points #Spacing of x values
dy = (ymax - ymin)/points #Spacing of y values

#Make an array to store the heights
z = zeros([points, points], float)

for i in range(points):
    y = ymin + i*dy

    for j in range(points):
        x = xmin + j*dx
        z[i,j] = cos(x)*sin(y)
        
#Make the plot
imshow(z, origin="lower", extent=[xmin, xmax, ymin, ymax], cmap="RdBu")
colorbar(label=r"$z(x,y)$")
xlabel("x (m)")
ylabel("y (m)")
show()
```

# Plotting p-orbitals of the hydrogen atom

```python
from numpy import zeros, sqrt, cos, sin, exp, arctan2, pi
from pylab import imshow, show, colorbar, xlabel, ylabel

def spherical(x, y):
    r = sqrt(x**2 + y**2)
    phi = arctan2(y, x)
    return r, phi

points = 500        # Number of grid points along each side
xmax = 10
xmin = -xmax
ymax = 10
ymin = -ymax

dx = (xmax - xmin)/points # Spacing of x values
dy = (ymax - ymin)/points # Spacing of y values

#Make an array to store wavefunction values
psi = zeros([points, points], float)

for i in range(points):
    y = ymin + i*dy

    for j in range(points):
        x = xmin + j*dx
        r, phi = spherical(x,y)
        psi[i,j] = 1/(4*sqrt(2*pi))*r*cos(phi)*exp(-r/2)
        
#Make the plot
imshow(psi, origin="lower", extent=[xmin, xmax, ymin, ymax], cmap="RdBu")
colorbar(label="$\psi_{2p_x}$")
xlabel("x")
ylabel("y")
show()
```

## Polar Plots

```python
import numpy as np
import pylab

def polar_to_cart(r, theta):
    x = r*np.cos(theta)
    y = r*np.sin(theta)
    return x, y

theta = np.linspace(0, 10*np.pi, 1000)
r = theta**2

x, y = polar_to_cart(r, theta)

pylab.plot(x, y, "r-", label=r"$r=\theta^2$")
pylab.legend(loc="upper right")
pylab.xlabel("x")
pylab.ylabel("y")
pylab.show()
```

```python

theta2 = np.linspace(0, 24*np.pi, 1000)
r2 = np.exp(np.cos(theta2)) - 2*np.cos(4*theta2) + (np.sin(theta2/12))**5
x, y = polar_to_cart(r2, theta2)
pylab.plot(x, y, "b-", label=r"$r=e^{cos(\theta)} - 2cos(4\theta) + sin^5 (\theta/12)$")
pylab.legend(loc="upper right")
pylab.xlabel("x")
pylab.ylabel("y")
pylab.show()


theta3 = np.linspace(0, 2*np.pi, 1000)
r3 = 2*(1+np.cos(theta3))
x, y = polar_to_cart(r3, theta3)
pylab.plot(x, y, "g--", label=r"$r=e^{cos(\theta)} - 2cos(4\theta) + sin^5 (\theta/12)$")
pylab.legend(loc="upper right")
pylab.xlabel("x")
pylab.ylabel("y")
pylab.show()
```

## Density Plots

```python
import numpy as np
import matplotlib.pyplot as plt

N = 500
x_grid = np.linspace(-10, 10, N)
y_grid = np.linspace(-10, 10, N)
z = np.zeros([N,N], float)

for i in range(len(y_grid)):
    for j in range(len(x_grid)):
        z[i,j] = np.cos(x_grid[j]) * np.sin(y_grid[i])

plt.imshow(z, origin="lower", extent=[-10,10,-10,10], cmap="plasma")
plt.colorbar(label=r"$z(x,y)")


```
