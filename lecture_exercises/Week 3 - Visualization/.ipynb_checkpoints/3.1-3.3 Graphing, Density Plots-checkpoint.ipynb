# Plotting a Function

```python
from pylab import plot, show
from numpy import linspace, sin #imports a new function - linspace

x = linspace(0,10,100) #makes an array 100 values that are evenly spaced from 0 to 10
y = sin(x) #creates an array of y values
plot(x,y)
show()
```

# Formatting a Plot

```python
from pylab import plot, show, ylim, xlabel, ylabel
from numpy import linspace, sin

x = linspace(0,10,100)
y = sin(x)
plot(x,y)
ylim(-1.3, 1.3)
xlabel("x")
ylabel("y = sin x")
show()
```

```python
plot(x,y, "go")
show()
```

```python
plot(x,y, "k--")
show()
```

```python
from pylab import plot, show, ylim, xlabel, ylabel, legend
from numpy import linspace, sin, cos

x = linspace(0,10,100)
y1 = sin(x)
y2 = cos(x)
plot(x,y1, label="sin x")
plot(x,y2, label="cos x")
ylim(-1.3, 1.3)
xlabel("x")
ylabel("y")
legend(loc="upper right")
show()
```

# Density Plots

```python
from numpy import zeros, sin, cos, pi
from pylab import imshow, show, colorbar, xlabel, ylabel

points = 500        # Number of grid points along each side
xmax = 10
xmin = -xmax
ymax = 10
ymin = -ymax

dx = (xmax - xmin)/points #Spacing of x values
dy = (ymax - ymin)/points #Spacing of y values

#Make an array to store the heights
z = zeros([points, points], float)

for i in range(points):
    y = ymin + i*dy

    for j in range(points):
        x = xmin + j*dx
        z[i,j] = cos(x)*sin(y)
        
#Make the plot
imshow(z, origin="lower", extent=[xmin, xmax, ymin, ymax], cmap="jet")
colorbar()
xlabel("x (m)")
ylabel("y (m)")
show()
```

```python
imshow(z, origin="lower", extent=[xmin, xmax, ymin, ymax], cmap="hot")
show()
```

```python
from numpy import meshgrid, linspace
from matplotlib.pyplot import contour
x_vals = linspace(-10, 10, 1000)
y_vals = linspace(-10, 10, 1000)
X, Y = meshgrid(x_vals, y_vals)

Z = cos(X)*sin(Y)

contour(X, Y, Z)
```
